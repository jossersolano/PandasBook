## Generacion de números pseudoaleatoria

```python
samples = np.random.standard_normal(size=(4, 4))
samples
```

```python
from random import normalvariate
N = 1_000_000
%timeit samples = [normalvariate(0, 1) for _ in range(N)]
```

```python
%timeit np.random.standard_normal(N)
```

```python
# esto es un ejemplo con IA porque percibo distinto el rendimiento con np.
# a pesar que me dice que el de numpy ejecuta otros procesos antes de hacer
# lo ordenado
from random import normalvariate
import numpy as np
import time

N = 1_000_000

# Python puro
start_py = time.time()
samples_py = [normalvariate(0, 1) for _ in range(N)]
end_py = time.time()
time_py = end_py - start_py

# NumPy
start_np = time.time()
samples_np = np.random.standard_normal(N)
end_np = time.time()
time_np = end_np - start_np

print(f"Tiempo Python puro: {time_py:.2f} segundos")
print(f"Tiempo NumPy: {time_np:.2f} segundos")

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

plt.hist(samples_py, bins=100, alpha=0.5, label='Python normalvariate')
plt.hist(samples_np, bins=100, alpha=0.5, label='NumPy standard_normal')

plt.legend()
plt.title("Distribución de números aleatorios")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

```python
nsteps = 1000
rng = np.random.default_rng(seed=12345)  # fresh random generator
draws = rng.integers(0, 2, size=nsteps)
steps = np.where(draws == 0, 1, -1)
walk = steps.cumsum()
```

```python
walk.min()
walk.max()
```

```python
(np.abs(walk) >= 10).argmax()
```

```python
nwalks = 5000
nsteps = 1000
draws = rng.integers(0, 2, size=(nwalks, nsteps)) # 0 or 1
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(axis=1)
walks
```

```python
walks.max()
walks.min()
```

```python
hits30 = (np.abs(walks) >= 30).any(axis=1)
hits30
hits30.sum() # Number that hit 30 or -30
```

```python
crossing_times = (np.abs(walks[hits30]) >= 30).argmax(axis=1)
crossing_times
```

```python
crossing_times.mean()
```
